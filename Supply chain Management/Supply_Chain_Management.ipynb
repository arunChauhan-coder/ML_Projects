## Data Preprocessing + EDA

## Missing Value Check

```python
df.isnull().sum()
```

```python
val = round(df['workers_num'].isnull().sum()/len(df)*100,2)
print(f"{val}% of workers_num value are missing")

val1 = round(df['wh_est_year'].isnull().sum()/len(df)*100,2)
print(f"{val1}% of wh_est_year value are missing")

val2 = round(df['approved_wh_govt_certificate'].isnull().sum()/len(df)*100,2)
print(f"{val2}% of approved_wh_govt_certificate value are missing")
```

- drop the wh_est_year column bcause imputing it will create bias
- Apply imputation technique for col workers_num and approved_wh_govt_certificate

```python
df.drop('wh_est_year',axis =1 ,inplace = True)
```

```python
#workers_num is a numeric col hence i used median imputation technqiue

df['workers_num'].fillna(df['workers_num'].median(),inplace= True)
df['workers_num'].isnull().sum()
```

```python
# approved_wh_govt_certificate is a cat column therefore using mode imputation approch
df['approved_wh_govt_certificate'].fillna(df['approved_wh_govt_certificate'].mode()[0],inplace=True)
df['approved_wh_govt_certificate'].isnull().sum()
```

## Spliting Data in dependend and independent variable

```python
x = df.drop('product_wg_ton',axis = 1)
y = df['product_wg_ton']
```

## Outlier Detection

## Splitting Data in Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=101)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

df_x_train = pd.DataFrame(data= x_train_sc,columns = x_train.columns)
df_x_test = pd.DataFrame(data= x_test_sc,columns = x_test.columns)
```
