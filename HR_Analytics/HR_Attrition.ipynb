## Encoding

```python
#Education, EnvironmentSatisfaction,JobInvolvement,JobLevel,JobSatisfaction, PerformanceRating,RelationshipSatisfaction,
#StockOptionLevel,WorkLifeBalance
#var have dtype as int but they have categorical features There fore converting it to Object
```

```python
df['Education']=df['Education'].astype('O')
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].astype('O')
df['JobInvolvement']=df['JobInvolvement'].astype('O')
df['JobLevel']=df['JobLevel'].astype('O')
df['JobSatisfaction']=df['JobSatisfaction'].astype('O')
df['PerformanceRating']=df['PerformanceRating'].astype('O')
df['RelationshipSatisfaction']=df['RelationshipSatisfaction'].astype('O')
df['StockOptionLevel']=df['StockOptionLevel'].astype('O')
df['WorkLifeBalance']=df['WorkLifeBalance'].astype('O')
```

```python
from sklearn.preprocessing import LabelEncoder
col=[]
label = LabelEncoder()
for i in df.columns:
    if df[i].dtype == 'O' or df[i].dtype == 'bool':
        if df[i].nunique() <3: # for bool
            
            df[i] = label.fit_transform(df[i])
        if df[i].nunique() > 2: # for multiple value
            col.append(i)
df = pd.get_dummies(df, columns= col,drop_first=True)
            
            
```

## Balance Data

```python
import imblearn
from imblearn.over_sampling import RandomOverSampler # SMOTE
ros = RandomOverSampler() # SMOTE()
x, y = ros.fit_resample(x, y)
print(x.shape, y.shape)
```

## Splitting in train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x, y, train_size=0.75, random_state=42)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

sc= StandardScaler()
x_train_sc = sc.fit_transform(x_train)
x_test_sc = sc.transform(x_test)

df_x_train = pd.DataFrame(data= x_train_sc,columns = x_train.columns)
df_x_test = pd.DataFrame(data= x_test_sc,columns = x_test.columns)
```

```python
df_x_train
```

## PCA

```python
df_x_train.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
x_train_pca = pca.fit_transform(df_x_train)
x_test_pca = pca.fit_transform(df_x_test)
```

```python
pd.DataFrame(x_train_pca).head()
```

```python
explained_variance = pca.explained_variance_ratio_
```

```python
explained_variance
```

## Model Building

### KNN Classifier without pca

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python

error = []

# Calculating error for K values between 1 and 40
for i in range(1, 40):  

    knn_classifier = KNeighborsClassifier(n_neighbors=i)
    knn_classifier.fit(df_x_train, y_train)
    pred_i = knn_classifier.predict(df_x_test)
    error.append(np.mean(pred_i != y_test.squeeze()))
```

```python

plt.figure(figsize=(12, 6))  
plt.plot(range(1, 40), error, color='red', linestyle='dashed', marker='o',  
         markerfacecolor='blue', markersize=10)
plt.title('Error Rate K Value')  
plt.xlabel('K Value')  
plt.ylabel('Mean Error')  
plt.show()
```

```python
# Appling the best k value
knn_classifier = KNeighborsClassifier(n_neighbors=7)  
knn_classifier.fit(df_x_train, y_train)  
y_pred_train = knn_classifier.predict(df_x_train)
y_pred_test = knn_classifier.predict(df_x_test)
```

```python
from sklearn.metrics import accuracy_score
print("Training Accuracy :", accuracy_score(y_train, y_pred_train))
print("****************"*5)
print("Test Accuracy :", accuracy_score(y_test, y_pred_test))
```

```python
from sklearn.model_selection import cross_val_score
training_accuracy = cross_val_score(knn_classifier, df_x_train, y_train, cv=10)
test_accuracy = cross_val_score(knn_classifier, df_x_test, y_test, cv=10)
print("Training Accuracy after CV :", training_accuracy.mean())
print("***********************"*5)
print("Test Accuracy after CV :", test_accuracy.mean())
```

## Support Vector Machine

```python
from sklearn.svm import SVC

#kernel - linear
svm_linear = SVC(kernel='linear')
svm_linear.fit(df_x_train, y_train)
y_pred_train_linear = svm_linear.predict(df_x_train)
y_pred_test_linear = svm_linear.predict(df_x_test)

#kernel - sigmoid
svm_sigmoid = SVC(kernel='sigmoid')
svm_sigmoid.fit(df_x_train, y_train)
y_pred_train_sigmoid = svm_sigmoid.predict(df_x_train)
y_pred_test_sigmoid = svm_sigmoid.predict(df_x_test)

#kernel - poly
svm_poly = SVC(kernel='poly')
svm_poly.fit(df_x_train, y_train)
y_pred_train_poly = svm_poly.predict(df_x_train)
y_pred_test_poly = svm_poly.predict(df_x_test)

#kernel - rbf
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(df_x_train, y_train)
y_pred_train_rbf = svm_rbf.predict(df_x_train)
y_pred_test_rbf = svm_rbf.predict(df_x_test)
```

```python
# Evaluation 

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print("Training Accuracy - Linear :", accuracy_score(y_train, y_pred_train_linear))
print("*************"*5)
print("Test Accuracy - Linear :", accuracy_score(y_test, y_pred_test_linear))
print("*************"*5)
print("Training Accuracy - sigmoid :", accuracy_score(y_train, y_pred_train_sigmoid))
print("*************"*5)
print("Test Accuracy - sigmoid :", accuracy_score(y_test, y_pred_test_sigmoid))
print("*************"*5)
print("Training Accuracy - poly :", accuracy_score(y_train, y_pred_train_poly))
print("*************"*5)
print("Test Accuracy - poly :", accuracy_score(y_test, y_pred_test_poly))
print("*************"*5)
print("Training Accuracy - rbf :", accuracy_score(y_train, y_pred_train_rbf))
print("*************"*5)
print("Test Accuracy - rbf :", accuracy_score(y_test, y_pred_test_rbf))
```
